```python
import pandas as pd
import matplotlib.pyplot as plt
import os

arquivos = {
    "Março": "SINASC_RO_2019_MAR.csv",
    "Abril": "SINASC_RO_2019_ABR.csv",
    "Maio": "SINASC_RO_2019_MAI.csv",
    "Junho": "SINASC_RO_2019_JUN.csv",
    "Dezembro": "SINASC_RO_2019_DEZ.csv"
}

os.makedirs("graficos_sinasc", exist_ok=True)

# Gerar gráficos para cada mês
for mes, arquivo in arquivos.items():
    df = pd.read_csv(arquivo)
    
    plt.figure(figsize=(8, 5))
    df["IDADEMAE"].hist(bins=20, edgecolor="black")
    plt.title(f"Distribuição da Idade das Mães em {mes}")
    plt.xlabel("Idade da Mãe")
    plt.ylabel("Frequência")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    
    # Salvar a figura
    plt.savefig(f"graficos_sinasc/{mes}_idade_maes.png")
    plt.show()
```
